```python
import matplotlib.pyplot as plt

labels = 'Injured', 'In a critical situation', 'died', 'survived'
sizes = [40, 15, 7, 38]
explode = (0.1, 0, 0, 0)  

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=220)
ax1.axis('equal')  

plt.show()
```

```python
import matplotlib.pyplot as plt 
labels = ['English', 'German', 'Russian']
values = [67,6,21]

bars = plt.bar(labels, values)
patterns = ['+','|','/']
for bar in bars:
    bar.set_hatch(patterns.pop())



plt.title('Persentage of languages around the world')
plt.xlabel('Language names',)
plt.ylabel('Percentage')


                  
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


labels = ['1990', '1994', '2001', '2004', '2014']
english = [10, 12, 24, 35, 60]
russian = [25, 32, 34, 20, 23]

x = np.arange(len(labels))  
width = 0.32 

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, english, width, label='English')
rects2 = ax.bar(x + width/2, russian, width, label='Russian')


ax.set_ylabel('Number of countries')
ax.set_title('Years of languages spoken by countries')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()



fig.tight_layout()

plt.show()
```

```python
x = [1, 3, 4, 5]
y = [2, 4, 6, 8]

plt.figure(figsize=(6,4),dpi = 100)

plt.plot(x,y, 'g^--', label = '2x')

x2 = np.arange(0,4.5, 0.5)
plt.plot(x2[:7], x2[:7]**2,'r', label = 'X^2')
plt.plot(x2[5:], x2[5:]**2,'r--')

plt.title("Our first silly chart", fontdict = {"fontname": "Times New Roman", "fontsize": 13 })
plt.xlabel("x values")
plt.ylabel("y values")

plt.xticks([0,1,2,3,4,])
plt.yticks([0,2,4,6,7,])

plt.legend()

plt.savefig('mysillygraph.png', dpi = 300 )
```
